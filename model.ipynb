```python
df=pd.read_csv(r'C:\Users\MSI PC\OneDrive\Documents\New folder (2)\csvf\archive\Car details v3.csv')
```

```python
df=df.drop(['torque','max_power'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.dropna(axis=0)
```

```python
df['mileage']=df['mileage'].str.split(' ').str.get(0)
df['engine']=df['engine'].str.split(' ').str.get(0)
```

```python
df['mileage']=df['mileage'].astype(float)
df['engine']=df['engine'].astype(float)
```

```python
df['mileage']=df['mileage'].astype(int)
df['engine']=df['engine'].astype(int)
df['seats']=df['seats'].astype(int)
```

```python
df.head()
```

```python
df['company']=df['name'].str.split(' ').str.get(0)
```

```python
df['name']=df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df.head()
```

```python
df.info()
```

```python
x=df.drop('selling_price',axis=1)
y=df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=2)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer

ohe=OneHotEncoder()
ohe.fit(x[['name','company','fuel','owner','seller_type','transmission']])

print(ohe.categories_)
ct=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel','owner','seller_type','transmission']),remainder='passthrough')
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()

from sklearn.pipeline import make_pipeline

pipe=make_pipeline(ct,lr)
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import pickle

pickle.dump(pipe,open('prediction.pkl','wb'))
```
